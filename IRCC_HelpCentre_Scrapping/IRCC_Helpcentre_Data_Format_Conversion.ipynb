# Converting IRCC Data to a New Format

```python
# Importing Libraries
import os
import json
from datetime import datetime
```

```python
# New format
data = {
    "Title": "IRCC Help Centre",
    "Source(link)": "https://www.cic.gc.ca/english/helpcentre/index-a-z-can.asp",
    "Date": datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    "subheadings": []
}
```

```python
# Converting data to new format
folder_path = 'C:/Users/Rabi33/Desktop/Toronto Chapter/toronto-canada-smartguide/src/data/IRCC_HelpCentre_Output'
# List all files in the folder
file_list = os.listdir(folder_path)
# Iterate through the files
for file_name in file_list[::2]:
    if os.path.isfile(os.path.join(folder_path, file_name)):
        path = f'{folder_path}/{file_name}'
        with open(path, 'r') as json_file:
            print(json_file)
            data_to_add = json.load(json_file)
        data['subheadings'].append({
            'subheading' : data_to_add['Title'],
            'Content' : data_to_add['Content']
        })

output_directory = "C:/Users/Rabi33/Desktop/Toronto Chapter/toronto-canada-smartguide/src/data/"
json_file_path = os.path.join(output_directory, 'IRCC Help Centre.json')

with open(json_file_path, 'w', encoding='utf-8') as json_file:
    json.dump(data, json_file, indent=4)
```

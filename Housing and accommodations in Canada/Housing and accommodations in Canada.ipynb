# Scraping Housing and accommodations in Canada: https://www.educanada.ca/study-plan-etudes/before-avant/housing-logement.aspx?lang=eng

```python
#importing libraries
import os
import requests
from bs4 import BeautifulSoup
import json
from datetime import datetime
import re
```

```python
url = "https://www.educanada.ca/study-plan-etudes/before-avant/housing-logement.aspx?lang=eng"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
section = soup.find('main', class_='container')
subsection_to_remove = section.find('div', class_='pull-right')
subsection_to_remove.extract()
tags = section.find_all(['h1', 'h2', 'h3', 'p', 'ul'])
```

```python
text_data = [{} for _ in range(10)]
between_h1_tags = 0
between_h2_tags = 0
between_h3_tags = 0
for tag in tags:
    if tag.name == 'h1':
        if tag.text == 'Housing and accommodations in Canada':
            between_h1_tags = 1
            text_data[0]={
                'subheading' : tag.text,
                'content' : ''
            }
        else:
            break
    elif tag.name == 'h2':
        if tag.text == 'Living on campus at college or university':
            between_h2_tags = 1
            between_h1_tags = 0
            between_h3_tags = 0
            text_data[1]={
                'subheading' : tag.text,
                'content' : ''
            }
        elif tag.text == 'Living off campus in private housing':
            between_h2_tags = 2
            between_h1_tags = 0
            between_h3_tags = 0
            text_data[2]={
                'subheading' : tag.text,
                'content' : ''
            }
    elif tag.name == 'h3':
        if tag.text == 'Residence room options':
            between_h3_tags = 1
            between_h2_tags = 1
            between_h1_tags = 0
            text_data[3]={
                'subheading' : tag.text,
                'content' : ''
            }
        elif tag.text == 'Benefits of residence':
            between_h3_tags = 2
            between_h2_tags = 1
            between_h1_tags = 0
            text_data[4]={
                'subheading' : tag.text,
                'content' : ''
            }
        elif tag.text == 'Housing terms in Canada':
            between_h3_tags = 3
            between_h2_tags = 2
            between_h1_tags = 0
            text_data[5]={
                'subheading' : tag.text,
                'content' : ''
            }
        elif tag.text == 'Signing a lease and paying a damage deposit':
            between_h3_tags = 4
            between_h2_tags = 2
            between_h1_tags = 0
            text_data[6]={
                'subheading' : tag.text,
                'content' : ''
            }
        elif tag.text == 'How to find private housing':
            between_h3_tags = 5
            between_h2_tags = 2
            between_h1_tags = 0
            text_data[7]={
                'subheading' : tag.text,
                'content' : ''
            }
        elif tag.text == 'Costs for private housing':
            between_h3_tags = 6
            between_h2_tags = 2
            between_h1_tags = 0
            text_data[8]={
                'subheading' : tag.text,
                'content' : ''
            }
        elif tag.text == 'Dealing with your landlord':
            between_h3_tags = 7
            between_h2_tags = 2
            between_h1_tags = 0
            text_data[9]={
                'subheading' : tag.text,
                'content' : ''
            }
    elif tag.name == 'p' or tag.name == 'ul':
        if between_h1_tags == 1:
            text_data[0]['content'] += tag.text
        elif between_h2_tags == 1 and between_h3_tags == 0:
            text_data[1]['content'] += tag.text
        elif between_h2_tags == 2 and between_h3_tags == 0:
            text_data[2]['content'] += tag.text
        elif between_h2_tags == 1 and between_h3_tags == 1:
            text_data[3]['content'] += tag.text
        elif between_h2_tags == 1 and between_h3_tags == 2:
            text_data[4]['content'] += tag.text
        elif between_h2_tags == 2 and between_h3_tags == 3:
            text_data[5]['content'] += tag.text
        elif between_h2_tags == 2 and between_h3_tags == 4:
            text_data[6]['content'] += tag.text
        elif between_h2_tags == 2 and between_h3_tags == 5:
            text_data[7]['content'] += tag.text
        elif between_h2_tags == 2 and between_h3_tags == 6:
            text_data[8]['content'] += tag.text
        elif between_h2_tags == 2 and between_h3_tags == 7:
            text_data[9]['content'] += tag.text

#get current directory
current_directory = os.getcwd()

# make output directory if not exist
output_directory = os.path.join(current_directory, "Output")
os.makedirs(output_directory, exist_ok=True)

json_data = {
    "Title": "Housing and accommodations in Canada",
    "Source(link)": f"{url}",
    "Date": datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    "subheadings": text_data
}

json_file_path = os.path.join(output_directory, f'Housing and accommodations in Canada.json')

with open(json_file_path, 'w', encoding='utf-8') as json_file:
    json.dump(json_data, json_file, indent=4)
```
